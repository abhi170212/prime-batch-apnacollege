```python
import requests;
from bs4 import BeautifulSoup;
URL="https://quotes.toscrape.com/"
response = requests.get(URL);
print(response.status_code) # for safety concerns
if response.status_code ==200:
        with open("dataStorage/lifeTagsData.txt","w",encoding="utf-8") as f:
            f.write(response.text);
else:
    print("data can not be taken as status code is " + response.status_code);
    

```

```python
with open("dataStorage/lifeTagsData.txt","r",encoding="utf-8") as f:
    html_data = f.read();
soup = BeautifulSoup(html_data,"lxml");
mainDiv = soup.find_all("div",class_="quote");
with open("dataStorage/QuotesOfTagLife.txt","w",encoding="utf-8")as f:
    for div in mainDiv:
        anchorTag= div.find_all("a",class_="tag");
        for tag in anchorTag:
            if( tag.get_text(strip=True) == "life"):
                data = div.get_text(strip=True);
                f.write(data + "\n\n");
        
```

```python
# pages contains pagination , then how to do it ?
import requests;
from bs4 import BeautifulSoup;
page_count = 1;
while True:
    URL = f"https://quotes.toscrape.com/page/{page_count}/";
    res =requests.get(URL);

    soup = BeautifulSoup(res.text,"lxml");
    quotes = soup.select("div.quote");
    if not quotes:
        print("no valid quotes");
        break;
    with open(f"scrapedPages/quotes{page_count}.html","w",encoding="utf-8") as f:
        f.write(res.text)
        print(f"Downloaded data from page{page_count}");
    page_count +=1;
```
