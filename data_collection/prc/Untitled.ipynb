```python
import requests;
from bs4 import BeautifulSoup;

URL = "https://www.scrapethissite.com/pages/simple/"
res = requests.get(URL);
if res.status_code == 200:
    print(len(res.text))
    with open("dataStorage/data.html","w",encoding="utf-8") as f:
        f.write(res.text);
else:
    print("There is an error in fetching ", res.status_code);
#####################################################################################


```

```python
# beautiful soup working 
# collect all the countries in a separate file 
from bs4 import BeautifulSoup;
with open("dataStorage/data.html","r") as f:
    html_content=f.read();
soup = BeautifulSoup(html_content,"lxml");
# countries list and store them in a file 
countries_list = soup.find_all("h3",class_="country-name")
# countries_list=countries_list.text;
with open("dataStorage/countriesList.txt","w",encoding="utf-8") as f:
    for country in countries_list:
        name=country.get_text(strip=True);
        f.write(name+ "\n" );

```

```python
# beautiful soup working 
# now collect countries and capital cities together 
from bs4 import BeautifulSoup;
with open("dataStorage/data.html","r") as f:
    html_content = f.read();
soup = BeautifulSoup(html_content,"lxml");
country_div = soup.find_all("div",class_="country");
with open("dataStorage/CountryCapitals.txt","w",encoding="utf-8")as f:
    for country in country_div:
        name = country.find("h3",class_="country-name").text;
        capital = country.find("span",class_="country-capital").get_text(strip=True);
        text = name + " " + capital;
        f.write(text+"\n");
        
```

```python
#bs working 
# countries with area and population 
from bs4 import BeautifulSoup;
with open("dataStorage/data.html","r")as f:
    html_content = f.read();
soup = BeautifulSoup(html_content,"lxml");
country_div = soup.find_all("div",class_="country");
with open("dataStorage/CountryAreaData.txt","w",encoding="utf-8") as f:
    for country in country_div:
        countryName = country.find("h3",class_="country-name").get_text(strip=True);
        countryArea=country.find("span",class_="country-area").get_text(strip=True);
        countryPopulation = country.find("span",class_="country-population").get_text(strip=True);
        
        data = "country "+countryName + "   Area "+countryArea + "  Population" + countryPopulation;
        f.write(data+"\n");
        
```
