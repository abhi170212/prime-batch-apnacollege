```python
#proeprties of series 
 # -> thy are homogenous 
s1 = pd.Series([3,2,1,6,3],index = ["adam","messi","luis",5,8])
print(s1.index.dtype) # aisa pta karte hai index ka dtype

#they have mutable values 
s3 = pd.Series([10, 20, 30])
s3[0] = 100
print(s3)

#they have immutable size 
s4 = pd.Series([10,20])
s4[2]= 60;
s4.drop(0)
print(s4)
```
