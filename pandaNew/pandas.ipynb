```python
#properties of series 
# 1) they are homongeneous 
ser2 = pd.Series([1,2],index=["a",4]); # same index ka type hota hai , as usual 
print(ser2)
```

```python
# 2) they support vectorization 
sr3 = pd.Series([1,4,6]);
sr4= pd.Series([10,40,10])
print(sr3+sr4)
```

```python
# 3) handling the missing values 
ser5 = pd.Series([1,2],index=["a","b"]);
```
