```python
#Box Plot --------------------------------------------------------------
import matplotlib.pyplot as plt
data =[7,8,6,9,4,10,12,15];
plt.boxplot(data);
plt.grid(True);
```

```python
#multiple datasets 
import numpy as np;
g1= np.random.normal(50,10,100);
g2=np.random.normal(60,15,100);
plt.boxplot([g1,g2],tick_labels=["group1","group2"]);
plt.grid(True);
plt.show()
```

```python
#horizontal box plot 
import matplotlib.pyplot as plt
data =[7,8,6,9,4,10,12,15];
plt.boxplot(data,vert=False);
plt.grid(True);
```

```python
#mean values of the data 
g1= np.random.normal(50,10,100);
g2=np.random.normal(60,15,100);
plt.boxplot([g1,g2],tick_labels=["group1","group2"],showmeans = True);
plt.grid(True);
plt.show()
```

```python
# Change the Whisker Length :-----------------------------------------------------
import matplotlib.pyplot as plt
data =[7,8,6,9,4,10,12,15];
plt.boxplot(data,vert=False,whis=2.0);
plt.grid(True);
```
