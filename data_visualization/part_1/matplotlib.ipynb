```python
 #real data 
oscar_movies = {
"The Dark Knight",
"The Hurt Locker",
"The King's Speech",
"The Artist",
"Argo"
}

years = [2008,2009,2010,2011,2012];
oscar_revenue = [1005,170,427,133,232];
#plot 
plt.plot(years,oscar_revenue);
plt.title("Oscar Movies Revenue in each Year");
plt.xlabel("years");
plt.ylabel("Revenue (in $M)");
```

```python
non_oscar_movies=["slumdog millionair","Avatar","Inception","Hugo","Lincoln"];
non_oscar_revenue=[378,2788,829,185,275]; # in $M

plt.plot(years,oscar_revenue,color="purple",marker="<",linewidth=5,linestyle="dashed",label="Oscar Movies");
plt.plot(years,non_oscar_revenue,"o--r",label="Non Oscar Movies");
plt.title("Oscar Movies vs Non-Oscar Movies Revenue");
plt.xlabel("Year");
plt.ylabel("Revenue (in $M)");
plt.legend(loc="upper right"); # this to show the label option 
```

```python
plt.plot(years,oscar_revenue,color="purple",label="Oscar Movies");
plt.plot(years,non_oscar_revenue,label="Non Oscar Movies");
plt.title("Oscar Movies vs Non-Oscar Movies Revenue");
plt.xlabel("Year");
plt.ylabel("Revenue (in $M)");
plt.legend(loc="upper right"); # this to show the label option 

print(plt.style.available);  
plt.style.use("dark_background");
plt.grid()
plt.show()
```
