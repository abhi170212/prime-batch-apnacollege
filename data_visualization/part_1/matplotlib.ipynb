```python
#bar graph 
import matplotlib.pyplot as plt
movies = [
"The Dark Knight",
"The Hurt Locker",
"The King's Speech",
"The Artist",
"Argo"
]

years = [2008,2009,2010,2011,2012];
revenue = [1005,170,427,133,232];
plt.bar(years,revenue,color="green")
plt.title('Revenue in each Year for oscar movies');
plt.xlabel("Year");
plt.ylabel("Revenue (in $M)");
# plt.text(years[0],revenue[0]+15,str(revenue[0]),ha="center") # agar single ko modify krna hai to
for i in range(len(revenue)):
    plt.text(years[i],revenue[i]+20,str(revenue[i]),ha="center");

plt.ylim(0,max(revenue)+100)
```

```python
#Grouped Bar Charts
import numpy as np;
non_oscar_movies=["slumdog millionair","Avatar","Inception","Hugo","Lincoln"];
non_oscar_revenue=[378,2788,829,185,275]; # in $M

oscar_movies = [
"The Dark Knight",
"The Hurt Locker",
"The King's Speech",
"The Artist",
"Argo"
]
oscar_revenue = [1005,170,427,133,232];

years = [2008,2009,2010,2011,2012];
years = np.array(years);
width=0.4;

plt.bar(years - width/2,oscar_revenue,width);
plt.bar(years + width/2,non_oscar_revenue,width)
plt.title("Oscar vs Non-Oscar Movies Revenue");
plt.ylabel("Revenue ($M)");
plt.legend(["Oscar Movies", "Non-Oscar Movies"])
plt.grid(axis="y", linestyle="--", alpha=0.6)
```

```python
#scatter plots 
people = [
    "Person A", "Person B", "Person C", "Person D", "Person E",
    "Person F", "Person G", "Person H", "Person I", "Person J"
]
age = [22, 25, 30, 35, 40, 45, 50, 55, 60, 65]
bp = [110, 115, 120, 122, 125, 130, 135, 123, 145, 150]
# colors = [
#     "crimson",
#     "royalblue",
#     "gold",
#     "mediumseagreen",
#     "darkorange",
#     "purple",
#     "deeppink",
#     "teal",
#     "slateblue",
#     "tomato"
# ]
colors=["green" if x<135 else "red" for x in bp];
# here one data is outliar cause his bp is normal with age 
plt.scatter(age,bp,color=colors,s=bp,marker="<",alpha=0.9);
plt.title("Age vs BP");
plt.xlabel("Age");
plt.ylabel("Blood Pressure");
plt.grid(True)
plt.colorbar(label="BP")

#annotation
for i in range(len(people)):
    plt.annotate(people[i],xy=(age[i],bp[i]),xytext=(age[i]+2,bp[i]+2))
plt.xlim(min(age),max(age)+13);   
plt.ylim(min(bp),max(bp)+8)
```

```python
import matplotlib.pyplot as plt

plt.style.use("default")   # style pehle

expenses = ["Salaries", "Rent", "Marketing", "R&D", "Miscellaneous"]
amounts = [500, 150, 120, 100, 50]
explode=[0,0,0,0.2,0];
plt.pie(amounts, labels=expenses,autopct="%1.1f%%",startangle=90,
    explode=explode,
        shadow=True,
    wedgeprops={
    "edgecolor":"black",
    "linewidth":1,
    "linestyle":"--",
    "fill":True , # color ko cehck krta hai ye 
})
plt.title("Company Expenses")
plt.show()
```
